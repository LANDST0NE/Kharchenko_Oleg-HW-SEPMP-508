Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-4 * np.pi, 4 * np.pi, 0.1)
y = np.sin(x) - 1

plt.plot(x, y)
plt.title("График синуса")
plt.xlabel("X")
plt.ylabel("SIN X")
plt.grid(True)
plt.show()
```

Задание 2

```python
amount = [26.98, 9.8, 9.65, 8.76, 4.87, 2.04, 15.96]
labels = ["Python", "C++", "C", "Java", "C#", "Golang", "COBOL"]
colors = plt.cm.viridis(np.linspace(0, 0.9, len(amount)))

plt.pie(amount, labels=labels, colors=colors, autopct='%1.f%%', labeldistance= 1.12)
plt.title("Самые популярные языки программирования")
plt.show()
```
